### 1. Simulate New Patient Check-in (User Input & `while` loop - Ch 7, 2, 3, 6)

* Use a `while` loop to continuously prompt the clinic staff (the user running the script) to enter information for *new* patients checking in.
* For each patient, collect their:
    * `name` (string)
    * `age` (integer - use `int()`)
    * `reason` for visit (string)
    * `arrival_time` (integer - minutes past opening).
* Store each new patient's data as a **dictionary**.
* Add each new patient dictionary to the `new_patient_entries` **list**.

````python
# Represents patient records already in the system
# We'll use simple integers for time just for this exercise (e.g., minutes past opening)
existing_patients = [
    {'name': 'John Smith', 'age': 45, 'reason': 'Checkup', 'arrival_time': 10, 'time_seen': 25, 'waiting_time': 15, 'category': 'Adult'},
    {'name': 'Maria Garcia', 'age': 8, 'reason': 'Fever', 'arrival_time': 15, 'time_seen': 35, 'waiting_time': 20, 'category': 'Pediatric'},
    {'name': 'David Lee', 'age': 68, 'reason': 'Blood Pressure', 'arrival_time': 20, 'time_seen': 30, 'waiting_time': 10, 'category': 'Senior'},
    {'name': 'Sophia Jones', 'age': 22, 'reason': 'Injury', 'arrival_time': 25, 'time_seen': 55, 'waiting_time': 30, 'category': 'Adult'}
]

# List to hold new patients entered during this session
new_patient_entries = []

active = True

while active:
  prompt_name = input("What is the patient's name? ")
  if prompt_name == "quit":
    active = False
  prompt_age = input("What is the patient's age? ")
  prompt_reason = input("What is the reason for the visit? ")
  prompt_arrival = input("What time did you arrive ")
  another_patient = input("Would you like to enter another patient (yes/no)")


  new_patients = {'name': prompt_name.title(), 'age': int(prompt_age), 'reason': prompt_reason, 'arrival_time': int(prompt_arrival)}

  new_patient_entries.append(new_patients)


  print(another_patient.lower())

  if another_patient != 'yes':
    active = False

print("You have completed entering the patient data:")

print(new_patient_entries)
````

### 2. Calculate Waiting Times (`for` loop, Dictionaries - Ch 4, 6)

* Imagine that *after* the check-in process, the system records the `time_seen` for the *newly entered* patients. For this exercise, *simulate* this by adding a `time_seen` value to each dictionary in the `new_patient_entries` list.

* Iterate through the `new_patient_entries` list using a `for` loop.
* For each patient dictionary, calculate the `waiting_time` (`time_seen` - `arrival_time`).
* Add a new key-value pair `'waiting_time': calculated_time` to that patient's dictionary.

````python
for patient in new_patient_entries:
  time_prompt = int(input(f"What time was {patient['name']} seen at: "))
  patient["time_seen"] = time_prompt

  waiting_time = patient['time_seen'] - patient['arrival_time']

  patient["waiting_time"] = waiting_time

print(new_patient_entries)
````

## Streamlit Code for web interfacing

````python
import streamlit as st

# Initialize session state for patient data
if 'patients' not in st.session_state:
    st.session_state['patients'] = []

st.title("Clinic EMR: Patient Waiting Time Analysis")

st.header("Add New Patient")
with st.form("patient_form"):
    name = st.text_input("Name")
    age = st.number_input("Age", min_value=0, max_value=120, step=1)
    reason = st.text_input("Reason for Visit")
    arrival_time = st.number_input("Arrival Time (minutes past opening)", min_value=0, step=1)
    time_seen = st.number_input("Time Seen (minutes past opening)", min_value=0, step=1)
    submitted = st.form_submit_button("Add Patient")
    if submitted:
        waiting_time = time_seen - arrival_time
        if age < 18:
            category = "Pediatric"
        elif age < 65:
            category = "Adult"
        else:
            category = "Senior"
        st.session_state['patients'].append({
            'name': name.title(),
            'age': age,
            'reason': reason,
            'arrival_time': arrival_time,
            'time_seen': time_seen,
            'waiting_time': waiting_time,
            'category': category
        })
        st.success(f"Added patient: {name.title()}")

st.header("Patient List")
if st.session_state['patients']:
    st.table(st.session_state['patients'])
else:
    st.write("No patients entered yet.")

st.header("Statistics")
if st.session_state['patients']:
    wait_times = [p['waiting_time'] for p in st.session_state['patients']]
    avg_wait = sum(wait_times) / len(wait_times)
    st.write(f"Average waiting time: {avg_wait:.2f} minutes")
    long_waits = [p for p in st.session_state['patients'] if p['waiting_time'] > 15]
    if long_waits:
        st.warning("Patients with long waits (>15 min):")
        for p in long_waits:
            st.write(f"{p['name']} waited {p['waiting_time']} minutes")
else:
    st.write("No statistics to show yet.")
````
